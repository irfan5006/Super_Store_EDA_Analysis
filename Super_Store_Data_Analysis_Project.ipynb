## __`Question No 06: üõí Are there seasonal trends in sales data?`__

```python
import statsmodels.api as sm
# Perform seasonal decomposition of sales data
decomposition = sm.tsa.seasonal_decompose(df['Sales'], model='additive', period=12)

# Create a seasonal decomposition plot with Seaborn
plt.figure(figsize=(18, 6))
sns.set_style("whitegrid")  # Optional: Set Seaborn style
sns.lineplot(data=decomposition.seasonal, label='Seasonal')
sns.lineplot(data=decomposition.trend, label='Trend')
sns.lineplot(data=decomposition.resid, label='Residuals')
plt.title('Seasonal Decomposition of Sales Data')
plt.xlabel('Time')
plt.legend()
plt.show()
```

### __Insight:__
- ‚ùå No apparent seasonal trend evident in the seasonal decomposition plot.
- üìä Sales data appears to have significant variability and noise.
- üßê Further analysis required to identify specific patterns or seasonal trends.
- üöÄ Insights from additional analysis may help in implementing effective strategies for sales improvement.-

## __`Question No 08: üèôÔ∏è Top 10  cities have the highest ?`__

```python

# Sort the cities by sales in descending order
city_sales_sorted = df.groupby('City')['Sales'].sum().reset_index()
city_sales_sorted = city_sales_sorted.sort_values(by='Sales', ascending=False)

# Select the top 10 cities with the highest sales
top_10_cities = city_sales_sorted.head(10)

# Create a line plot for the top 10 cities with the highest sales
plt.figure(figsize=(18, 6))
sns.lineplot(data=top_10_cities, x='City', y='Sales')
plt.title('Top 10 Cities with Highest Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Rotate city names for better visibility
plt.show()

```

## __Insight:__
- üåÜ The top 10 cities with the highest sales are led by `New York` City, followed by Los Angeles and `San Francisco.`
Recommendation:

- üöÄ Consider focusing resources and strategies on these `high-performing cities` to further boost sales and market presence.

## __`Question 11: üìÖ How do sales vary by order date?`__

```python
# Convert the 'Order Date' column to a datetime data type
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Calculate total sales by order date
sales_by_date = df.groupby('Order Date')['Sales'].sum().reset_index()

# Create a time series plot for sales by order date
plt.figure(figsize=(18, 6))
sns.lineplot(data=sales_by_date, x='Order Date', y='Sales')
plt.title('Sales Variation by Order Date (Time Series)')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate date labels for better visibility
plt.show()
```

## __` Question 12: üìà How does the quantity of products purchased change over time? `__

```python
# Convert the 'Order Date' column to a datetime data type
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Calculate total quantity by order date
quantity_by_date = df.groupby('Order Date')['Quantity'].sum().reset_index()

# Create a time series plot for quantity by order date with customizations
plt.figure(figsize=(18, 6))
sns.set(style="whitegrid")  # Set the style
sns.lineplot(data=quantity_by_date, x='Order Date', y='Quantity', color='blue')  # Change line color
plt.title('Quantity of Products Purchased Over Time (Time Series)', fontsize=16, color='green')  # Change title font size and color
plt.xlabel('Order Date', fontsize=12, color='purple')  # Change x-axis label font size and color
plt.ylabel('Total Quantity', fontsize=12, color='orange')  # Change y-axis label font size and color
plt.xticks(rotation=45)  # Rotate date labels for better visibility
plt.show()

```

## __`Question No 13: üìà How does profit change over the years? `__

```python

# Extract the year from the 'Order Date' column
df['Year'] = df['Order Date'].dt.year

# Calculate total profit by year
profit_by_year = df.groupby('Year')['Profit'].sum().reset_index()

# Create a line chart for profit by year
plt.figure(figsize=(18, 6))
sns.lineplot(x='Year', y='Profit', data=profit_by_year)
plt.title('Profit Variation Over the Years (Line Chart)')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.show()
```
